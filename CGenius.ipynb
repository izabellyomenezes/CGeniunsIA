#### Importando Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Visualização tabela

```python
clientes = pd.read_csv('clientes.csv')
atendentes = pd.read_csv('atendentes.csv')
```

```python
clientes.info()
```

```python
atendentes.info()
```

#### Tratando e organizando dados 

```python
clientes.columns = ['fatura', 'id', 'genero', 'idade', 'categoria', 'quantidade', 'preco', 'metodoPagamento', 'dataCompra', 'shopping'] 
atendentes.columns = ['educação', 'anoDeEntrada', 'cidade', 'categoriaPagamento', 'idade', 'genero', 'ficouNoBanco', 'tempoExperiencia', 'saiuOuNao' ]
```

```python
clientes.head()
atendentes.head()
```

```python
clientes.dropna 
atendentes.dropna
```

```python
clientes.isna()
atendentes.isna()
```

#### Visualização de dados 

```python
plt.figure(figsize=(10, 6))
sns.histplot(clientes['idade'], bins=30)
plt.title('Idades dos Clientes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(atendentes['idade'], bins=30)
plt.title('Idades dos atendentes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()

plt.figure(figsize=(15, 3))
sns.countplot(x='idade', data=clientes, label='idade', alpha=0.7)
sns.countplot(x='idade', data=atendentes, label='idade', alpha=0.7)
plt.title('Média')
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(clientes['genero'], bins=30) 
plt.title('genero dos Clientes')
plt.xlabel('genero')
plt.ylabel('Frequência')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(atendentes['genero'], bins=30)
plt.title('genero dos atendentes')
plt.xlabel('genero')
plt.ylabel('Frequência')
plt.show()

plt.figure(figsize=(5, 4))
sns.countplot(x='genero', data=clientes, label='Clientes', alpha=0.7)
sns.countplot(x='genero', data=atendentes, label='Atendentes', alpha=0.7)
plt.title('Gênero')
plt.legend()
plt.show()
```
